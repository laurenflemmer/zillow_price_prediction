## Creating distance variables
### Distance from downtown LA

```python
# fix lat, long values in data
adj_coords = train_df[["latitude", "longitude"]]/1e6

dtla_coords = (34.0488, -118.2518)
property_coords = adj_coords.values.tolist()
property_coords = [tuple(x) for x in property_coords]


# calculate haversine distance from each property to downtown la
train_df["dist_to_dtla"] = list(map(lambda coord: haversine(coord, dtla_coords, unit='mi'), property_coords))
```

### Distance to the Ocean

```python
# create coordinate list for points along the coast
coast_data = pd.read_csv("/accounts/grad/huz4/Desktop/STAT222/Zillow-project/coast_data.csv")

# find point where the latitude of the property is equal to the latitude of the coastline (closest distance to the ocean)
closest_pair_list = []
for lat in adj_coords["latitude"]:
    lat_dist_list = abs(coast_data["Y"] - lat)
    min_lat_dist_idx = (np.array(lat_dist_list)).argmin()

    # get corresponding (lat, long) pair for the coastline point
    closest_pair_all = coast_data.iloc[min_lat_dist_idx]
    closest_pair_list.append(closest_pair_all[["Y", "X"]])
```

```python
# convert to tuple
closest_pair_list = [tuple(x) for x in closest_pair_list]
print(closest_pair_list[0:10])

# compute haversine distance from property to the nearest coastline location
dist_ocean_list = []
for i in range(len(property_coords)):
    dist_ocean_list.append(haversine(property_coords[i], closest_pair_list[i], unit='mi'))

train_df["dist_to_ocean"] = dist_ocean_list
train_df.head()
```
